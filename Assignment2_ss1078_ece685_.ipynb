## Problem 1: Conditional Multivariate Gaussian Distribution

**Part a: 
Find conditional PDF of $x_{2}$ given $x_{1}$ = a where a is a constant vector**

By definition we have: 

$$\left[\begin{array}{l}
{x_{1}} \\
{x_{2}}
\end{array}\right] \sim \mathcal{N}\left(\left[\begin{array}{l}
\mu_{x_{1}} \\
\mu_{x_{2}}
\end{array}\right],\left[\begin{array}{ll}
\Sigma_{x_{1} x_{1}} & \Sigma_{x_{1} x_{2}} \\
\Sigma_{x_{2} x_{1}} & \Sigma_{x_{2} x_{2}}
\end{array}\right]\right)$$

Then, the marginal densities are also a Gaussian distribution:

$$p\left(x_{1}\right)=\int_{x_{2} \in \mathbf{R}^{p}} p\left(x_{1}, x_{2} ; \mu, \Sigma\right) d x_{2}$$
$$p\left(x_{2}\right)=\int_{x_{1} \in \mathbf{R}^{D-p}} p\left(x_{1}, x_{2} ; \mu, \Sigma\right) d x_{1}$$
or:
$$
\begin{array}{l}
x_{1} \sim \mathcal{N}\left(\mu_{x_{1}}, \Sigma_{x_{1} x_{1}}\right) \\
x_{2} \sim \mathcal{N}\left(\mu_{x_{2}}, \Sigma_{x_{2} x_{2}}\right)
\end{array}
$$

By Bayes Theorem, we have:

$$
\begin{array}{l}
p(x_{2} \mid x_{1}) = \frac{p(x_{1},x_{2})}{p(x_{1})}&\propto p(x_{1},x_{2})
\end{array}
$$
This is because $p(x_{1})$ can be considered a constant. Now, by definition of conditional distribution for MVN:

$$
p\left(x_{2} \mid x_{1}\right) \propto \exp \left(-\frac{1}{2}\left[\begin{array}{l}
x_{1}-\mu_{x_{1}} \\
x_{2}-\mu_{x_{2}}
\end{array}\right]^{T}\left[\begin{array}{ll}
\Sigma_{x_{1} x_{1}} & \Sigma_{x_{1} x_{2}} \\
\Sigma_{x_{2} x_{1}} & \Sigma_{x_{2} x_{2}}
\end{array}\right]^{-1}\left[\begin{array}{l}
x_{1}-\mu_{x_{1}} \\
x_{2}-\mu_{x_{2}}
\end{array}\right]\right)$$



Let us define a matrix V in terms of covariances such that:
$$
V=\left[\begin{array}{ll}
V_{x_{1} x_{1}} & V_{x_{1} x_{2}} \\
V_{x_{2} x_{1}} & V_{x_{2} x_{2}}
\end{array}\right]=\left[\begin{array}{ll}
\Sigma_{x_{1} x_{1}} & \Sigma_{x_{1} x_{2}} \\
\Sigma_{x_{2} x_{1}} & \Sigma_{x_{2} x_{2}}
\end{array}\right]^{-1}
$$


On expanding:
$$
V=\left[\begin{array}{cc}
\left.\sum_{x_{1} x_{1}}-\Sigma_{x_{1} x_{2}} \Sigma_{x_{2} x_{2}}^{-1} \Sigma_{x_{2} x_{1}}\right)^{-1} & -\left(\Sigma_{x_{1} x_{1}}-\Sigma_{x_{1} x_{2}} \Sigma_{x_{2} x_{2}}^{-1} \Sigma_{x_{2} x_{1}}\right)^{-1} \Sigma_{x_{1} x_{2}} \Sigma_{x_{2} x_{2}}^{-1} \\
-\Sigma_{x_{2} x_{2}}^{-1} \Sigma_{x_{2} x_{1}}\left(\Sigma_{x_{1} x_{1}}-\Sigma_{x_{1} x_{2}} \Sigma_{x_{2} x_{2}}^{-1} \Sigma_{x_{2} x_{1}}\right)^{-1} & \left(\Sigma_{x_{2} x_{2}}-\Sigma_{x_{2} x_{1}} \Sigma_{x_{1} x_{1}}^{-1} \Sigma_{x_{1} x_{2}}\right)^{-1}
\end{array}\right]
$$
Using Sherman–Morrison–Woodbury formula, we can establish the relationship between the covariance matrices and $V$ like this:.


$$
\left[\begin{array}{cc}
\Sigma_{x_{1} x_{1}} & \Sigma_{x_{1} x_{2}} \\
\Sigma_{x_{2} x_{1}} & \Sigma_{x_{2} x_{2}}
\end{array}\right]=\left[\begin{array}{cc}
\left(V_{x_{1} x_{1}}-V_{x_{1} x_{2}} V_{x_{2} x_{2}}^{-1} V_{x_{2} x_{1}}\right)^{-1} & -\left(V_{x_{1} x_{1}}-V_{x_{1} x_{2}} V_{x_{2} x_{2}}^{-1} V_{x_{2} x_{1}}\right)^{-1} V_{x_{1} x_{2}} V_{x_{2} x_{2}}^{-1} \\
-V_{x_{2} x_{2}}^{-1} V_{x_{2} x_{1}}\left(V_{x_{1} x_{1}}-V_{x_{1} x_{2}} V_{x_{2} x_{2}}^{-1} V_{x_{2} x_{1}}\right)^{-1} & \left(V_{x_{2} x_{2}}-V_{x_{2} x_{1}} V_{x_{1} x_{1}}^{-1} V_{x_{1} x_{2}}\right)^{-1}
\end{array}\right]$$

Plugging this into the expression for $p(x_{2}\mid x_{1})$:

$$
\begin{array}{c}
p\left(x_{2} \mid x_{1}\right) \propto \exp \left(-\frac{1}{2}\left[\begin{array}{c}
x_{1}-\mu_{x_{1}} \\
x_{2}-\mu_{x_{2}}
\end{array}\right]^{T}\left[\begin{array}{cc}
V_{x_{1} x_{1}} & V_{x_{1} x_{2}} \\
V_{x_{2} x_{1}} & V_{x_{2} x_{2}}
\end{array}\right]\left[\begin{array}{c}
x_{1}-\mu_{x_{1}} \\
x_{2}-\mu_{x_{2}}
\end{array}\right]\right) \\
\propto \exp \left(-\left[\frac{1}{2}\left(x_{1}-\mu_{x_{1}}\right)^{T} V_{x_{1} x_{1}}\left(x_{1}-\mu_{x_{1}}\right)+\frac{1}{2}\left(x_{1}-\mu_{x_{1}}\right)^{T} V_{x_{1} x_{2}}\left(x_{2}-\mu_{x_{2}}\right)+\frac{1}{2}\left(x_{2}-\mu_{x_{2}}\right)^{T} V_{x_{2} x_{1}}\left(x_{1}-\mu_{x_{1}}\right)+\frac{1}{2}\left(x_{2}-\mu_{x_{2}}\right)^{T} V_{x_{2} x_{2}}\left(x_{2}-\mu_{x_{2}}\right)\right]\right)
\end{array}
$$

After completing the squares, we get:

$$
\left.\begin{array}{rl}
p\left(x_{2} \mid x_{1}\right) \propto \exp \left(-\left[\frac{1}{2}\left(x_{2}-\mu_{x_{2}}+V_{x_{2} x_{2}}^{-1} V_{x_{2} x_{1}}\left(x_{1}-\mu_{x_{1}}\right)\right)^{T} V_{x_{2} x_{2}}\left(x_{2}-\mu_{x_{2}}+V_{x_{2} x_{2}}^{-1} V_{x_{2} x_{1}}\left(x_{1}-\mu_{x_{1}}\right)\right)\right.\right. \\
\left.+\frac{1}{2}\left(x_{1}-\mu_{x_{1}}\right)^{T} V_{x_{1} x_{1}}\left(x_{1}-\mu_{x_{1}}\right)-\frac{1}{2}\left(x_{1}-\mu_{x_{1}}\right)^{T} V_{x_{1} x_{2}} V_{x_{2} x_{2}}^{-1} V_{x_{2} x_{1}}\left(x_{1}-\mu_{x_{1}}\right)\right]
\end{array}\right)$$

From this, it follows that the conditional densities take the following form:

$$p\left(x_{2} \mid x_{1}\right)=\frac{p\left(x_{2}, x_{1} ; \mu, \Sigma\right)}{\int_{x_{2} \in \mathbf{R}^{D-p}} p\left(x_{2}, x_{1} ; \mu, \Sigma\right) d x_{2}}$$

Thus, $$p\left(x_{2} \mid x_{1}\right)=\mathcal{N}\left(x_{2} \mid \mu_{x_{2} \mid x_{1}}, \Sigma_{x_{2} \mid x_{1}}\right)$$

This is a Gaussian Distribution.
Now, for $\mathcal{N}\left(x_{2} \mid \mu_{x_{2} \mid x_{1}}, \Sigma_{x_{2} \mid x_{1}}\right)$, $\mu_{x_{2} \mid x_{1}}$ and $\Sigma_{x_{2} \mid x_{1}}$ can be computed in the following manner:


$$\boldsymbol{\Sigma}_{a \mid b}=\boldsymbol{\Lambda}_{a a}^{-1}=\boldsymbol{\Sigma}_{a a}-\boldsymbol{\Sigma}_{a b} \boldsymbol{\Sigma}_{b b}^{-1} \boldsymbol{\Sigma}_{b a}$$ (from class slides)

Thus, in this case: 

$$\boldsymbol{\Sigma}_{x_{2} \mid x_{1}}=\boldsymbol{\Lambda}_{x_{2} x_{2}}^{-1}=\boldsymbol{\Sigma}_{x_{2} x_{2}}-\boldsymbol{\Sigma}_{x_{2} x_{1}} \boldsymbol{\Sigma}_{x_{1} x_{1}}^{-1} \boldsymbol{\Sigma}_{x_{1} x_{2}}$$


For the mean:
$$
\boldsymbol{\mu}_{a \mid b}=\boldsymbol{\Sigma}_{a \mid b}\left\{\boldsymbol{\Lambda}_{a a} \boldsymbol{\mu}_{a}-\boldsymbol{\Lambda}_{a b}\left(\mathbf{x}_{b}-\boldsymbol{\mu}_{b}\right)\right\}
$$

Thus,
$$\boldsymbol{\mu}_{x_{2} \mid x_{1}}=\boldsymbol{\Sigma}_{x_{2} \mid x_{1}}\left\{\boldsymbol{\Lambda}_{x_{2} x_{2}} \boldsymbol{\mu}_{x_{2}}-\boldsymbol{\Lambda}_{x_{2} x_{1}}\left(\mathbf{x}_{x_{1}}-\boldsymbol{\mu}_{x_{1}}\right)\right\}$$


$$\begin{array}{l}
\boldsymbol{\mu}_{x_{2} \mid x_{1}}=\boldsymbol{\mu}_{x_{2}}-\boldsymbol{\Lambda}_{x_{2} x_{2}}^{-1} \boldsymbol{\Lambda}_{x_{2} x_{1}}\left(\mathbf{x}_{x_{1}}-\boldsymbol{\mu}_{x_{1}}\right) \\
\boldsymbol{\mu}_{x_{2} \mid x_{1}}=\boldsymbol{\mu}_{x_{2}}+\boldsymbol{\Sigma}_{x_{2} x_{1}} \boldsymbol{\Sigma}_{x_{1} x_{1}}^{-1}\left(\mathbf{x}_{x_{1}}-\boldsymbol{\mu}_{x_{1}}\right)
\end{array}$$

## Problem 2: Gaussian Mixture Model

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
def gmm_distr(n,param):
    '''
    function to generate a random gmm distribution
    '''
    cov,mu = param
    return np.random.multivariate_normal(mu,cov,n).T

def gen_samples(n,distr,covs,mus,ps):
    '''
    function to generate n samples from a given distribution function
    '''
    random_choice = np.random.choice(len(ps),n,replace=True, p=ps)
    num_samples = np.bincount(random_choice)
    samples = []
    for idx in range(len(ps)):
        param = (covs[idx],mus[idx])
        samples.append(distr(num_samples[idx],param))
        pass
    return np.concatenate(samples,axis=1)

def gmm_model(X,k_values):
    '''
    function to fit an sklearn Gaussian Mixture Model and find log likelihood scores
    '''
    scores = []
    for k in k_values:
        gmm_mod = GaussianMixture(n_components=k,random_state=20)
        gmm_mod = gmm_mod.fit(X)
        scores.append(gmm_mod.score(X))
    return scores

#define covariance matrices
covs = [ np.array([[0.2,0.1],[0.1,0.2]]),  np.array([[0.1,-0.15],[-0.15,0.3]]), 
        np.array([[0.3,0.05],[0.05,0.3]]), np.array([[0.15,0],[0,0.15]])]
#define means
mus = [np.array([0,0]),np.array([0,2]),np.array([2,0]),np.array([2,2])]
#define pi values
pis = [1/8,1/8,1/4,1/2]
#define K values
K_array = [1,2,3,4,5,6,7]

#generate 10^3 2-dimensional random samples from GMM distribution
n = 10**3
X = gen_samples(n,gmm_distr,covs,mus,pis)

#get log likelihood scores for plotting
scores = gmm_model(X.T,K_array)

########## plots ########### 
fig, ax = plt.subplots(1,2,figsize=(15,5))
#plot scatter plot of the generated samples
ax[0].scatter(X[0,:],X[1,:])
ax[0].set_title('D-Dimensional Gaussian Mixture Model')
ax[0].set_xlabel('x') 
ax[0].set_ylabel('y')


#plot the log likelihood
ax[1].plot(K_array,scores)
ax[1].set_title('GMM: Log Likelihood versus number of components')
ax[1].set_xlabel('Number of components') 
ax[1].set_ylabel('Log Likelihood')
plt.show()
```

## Problem 3: Bayesian linear regression

**Part a: What is the distribution of coordinates of the object after observing the sensor measurements? Please provide all the details including the pdf, and all the necessary moments information.**

The distribution would be a **Multivariate Normal Distribution (MVN).**

We have $x \sim \mathcal{N}\left(\mu_{0}, \Sigma_{0}\right)$. Using this, we can define the prior distribution as follows:

$$
\begin{aligned}
\pi(\boldsymbol{x}) &=(2 \pi)^{-\frac{p}{2}}\left|\Sigma_{0}\right|^{-\frac{1}{2}} \exp \left\{-\frac{1}{2}\left(\boldsymbol{x}-\boldsymbol{\mu}_{0}\right)^{T} \Sigma_{0}^{-1}\left(\boldsymbol{x}-\boldsymbol{\mu}_{0}\right)\right\} \\
& \propto \exp \left\{-\frac{1}{2}\left(\boldsymbol{x}-\boldsymbol{\mu}_{0}\right)^{T} \Sigma_{0}^{-1}\left(\boldsymbol{x}-\boldsymbol{\mu}_{0}\right)\right\}
\\
& \propto \exp \left\{-\frac{1}{2}\left[\boldsymbol{x}^{T}\Sigma_{0}^{-1}\boldsymbol{x}-\boldsymbol{x}^{T}\Sigma_{0}^{-1}\boldsymbol{\mu}_{0}-\boldsymbol{\mu}_{0}^{T}\Sigma_{0}^{-1}\boldsymbol{x}+\boldsymbol{\mu}_{0}^{T}\Sigma_{0}^{-1}\boldsymbol{\mu_{0}}      \right]\right\}
\end{aligned}
$$
Combining the second and third term above and removing the last term (with $\mu_{0}$ as it is a constant), we get:


$$
\begin{aligned}
\pi(\boldsymbol{x}) & \propto \exp \left\{-\frac{1}{2}\left[\boldsymbol{x}^{T}\Sigma_{0}^{-1}\boldsymbol{x}-2 \boldsymbol{x}^{T}\Sigma_{0}^{-1} \boldsymbol{\mu}_{0} \right]\right\} \\
 & \propto \exp \left\{-\frac{1}{2}\boldsymbol{x}^{T}\Sigma_{0}^{-1}\boldsymbol{x}+ \boldsymbol{x}^{T}\Sigma_{0}^{-1} \boldsymbol{\mu}_{0} \right\} 
\end{aligned}
$$

Now, we also have been given $y_{i} \sim \mathcal{N}\left(x, \Sigma_{y}\right)$ for $i \in (1,N)$. So we can find the likelihood as:
$$
\begin{array}{c}
L(\boldsymbol{Y} ; \boldsymbol{x}, \Sigma_{y}) \propto|\Sigma_{y}|^{-\frac{N}{2}} \exp \left\{-\frac{1}{2} \sum_{i=1}^{N}\left(\boldsymbol{y}_{i}-\boldsymbol{x}\right)^{T} \Sigma_{y}^{-1}\left(\boldsymbol{y}_{i}-\boldsymbol{x}\right)\right\} \\
\propto \exp \left\{-\frac{1}{2} \sum_{i=1}^{N}\left(\boldsymbol{y}_{i}^{T}-\boldsymbol{x}^{T}\right) \Sigma_{y}^{-1}\left(\boldsymbol{y}_{i}-\boldsymbol{x}\right)\right\} \\
\propto \exp \left\{-\frac{1}{2} \sum_{i=1}^{N}\left[\boldsymbol{y}_{i}^{T}\Sigma_{y}^{-1}\boldsymbol{y}_{i}-
\boldsymbol{y}_{i}^{T}\Sigma_{y}^{-1}\boldsymbol{x}-\boldsymbol{x}^{T}\Sigma_{y}^{-1}\boldsymbol{y}_{i}   +\boldsymbol{x}^{T}\Sigma_{y}^{-1}\boldsymbol{x}\right]\right\}
\end{array}
$$

On combining the second and third terms and removing the final term, we get:

$$
\begin{array}{l}
L(\boldsymbol{Y} ; \boldsymbol{x}, \Sigma_{y}) \propto \exp \left\{-\frac{1}{2} \sum_{i=1}^{N}\left[\boldsymbol{x}^{T} \Sigma_{y}^{-1} \boldsymbol{x}-2 \boldsymbol{x}^{T} \Sigma_{y}^{-1} \boldsymbol{y}_{i}\right]\right\} \\
\propto \exp \left\{-\frac{1}{2} \sum_{i=1}^{N} \boldsymbol{x}^{T} \Sigma_{y}^{-1} \boldsymbol{x}-\frac{1}{2} \sum_{i=1}^{N}(-2) \boldsymbol{x}^{T} \Sigma_{y}^{-1} \boldsymbol{y}_{i}\right\} \\
\propto \exp \left\{-\frac{1}{2} N \boldsymbol{x}^{T} \Sigma_{y}^{-1} \boldsymbol{x}+\boldsymbol{x}^{T} \Sigma_{y}^{-1} \sum_{i=1}^{N} \boldsymbol{y}_{i}\right\} \\
\propto \exp \left\{-\frac{1}{2} \boldsymbol{x}^{T}\left(N \Sigma_{y}^{-1}\right) \boldsymbol{x}+\boldsymbol{x}^{T}\left(N \Sigma_{y}^{-1} \overline{\boldsymbol{y}}\right)\right\}
\end{array}
$$

Here, $N*\overline{\boldsymbol{y}}= \Sigma_{y}$.

Now, since $\pi(\boldsymbol{x}|\Sigma_{y},Y) \propto L(\boldsymbol{Y} ; \boldsymbol{x}, \Sigma_{y}) \pi(\boldsymbol{x})$, we can derive the full conditional in the following manner:

$$
\begin{array}{l}
\pi(\boldsymbol{x}|\Sigma_{y},Y) \propto L(\boldsymbol{Y} ; \boldsymbol{x}, \Sigma_{y}) \pi(\boldsymbol{x}) \\
\propto \exp \left\{-\frac{1}{2} \boldsymbol{x}^{T}\left(N \Sigma_{y}^{-1}\right) \boldsymbol{x}+\boldsymbol{x}^{T}\left(N \Sigma_{y}^{-1} \overline{\boldsymbol{y}}\right)\right\}.\exp \left\{-\frac{1}{2}\boldsymbol{x}^{T}\Sigma_{0}^{-1}\boldsymbol{x}+ \boldsymbol{x}^{T}\Sigma_{0}^{-1} \boldsymbol{\mu}_{0} \right\} 
\end{array}
$$
 
On rearranging and clubbing the terms above:
$$
\begin{array}{l}
\pi(\boldsymbol{x}|\Sigma_{y},Y) \propto \exp \left\{
-\frac{1}{2} \boldsymbol{x}^{T}\left(N \Sigma_{y}^{-1}\right) \boldsymbol{x}  -\frac{1}{2}\boldsymbol{x}^{T}\Sigma_{0}^{-1}\boldsymbol{x} +\boldsymbol{x}^{T}\left(N \Sigma_{y}^{-1} \overline{\boldsymbol{y}}\right)  + \boldsymbol{x}^{T}\Sigma_{0}^{-1} \boldsymbol{\mu}_{0}\right\} \\
\propto \exp \left\{
-\frac{1}{2} \boldsymbol{x}^{T}\left[N \Sigma_{y}^{-1} +\Sigma_{0}^{-1}\right] \boldsymbol{x}  +\boldsymbol{x}^{T}\left[N \Sigma_{y}^{-1} \overline{\boldsymbol{y}}+\Sigma_{0}^{-1} \boldsymbol{\mu}_{0}\right] \right\}
\end{array}
$$

The above distribution is another MVN with the following moments:

$$
\begin{array}{l}
a_{n}=\left[N \Sigma_{y}^{-1} +\Sigma_{0}^{-1}\right] \\
Covariance= \Sigma_{n}= a_{n}^{-1}=\left[N \Sigma_{y}^{-1} +\Sigma_{0}^{-1}\right]^{-1}\\
Mean= \mu_{n}= a_{n}^{-1}\left[N \Sigma_{y}^{-1} \overline{\boldsymbol{y}}+\Sigma_{0}^{-1} \boldsymbol{\mu}_{0}\right] = \left[N \Sigma_{y}^{-1} +\Sigma_{0}^{-1}\right]^{-1}\left[N \Sigma_{y}^{-1} \overline{\boldsymbol{y}}+\Sigma_{0}^{-1} \boldsymbol{\mu}_{0}\right]=  \left[N \Sigma_{y}^{-1} +\Sigma_{0}^{-1}\right]^{-1}\left[ \Sigma_{y}^{-1} \sum_{i=1}^{N} y_{i}+\Sigma_{0}^{-1} \boldsymbol{\mu}_{0}\right]
\end{array}
$$

**Part b: What is your predicted location based on the sensor measurements? What is the MLE estimate?**

The predicted location on the basis of the sensor measurements is the posterior mean:

$$
\left[N \Sigma_{y}^{-1} +\Sigma_{0}^{-1}\right]^{-1}\left[N \Sigma_{y}^{-1} \overline{\boldsymbol{y}}+\Sigma_{0}^{-1} \boldsymbol{\mu}_{0}\right]=  \left[N \Sigma_{y}^{-1} +\Sigma_{0}^{-1}\right]^{-1}\left[ \Sigma_{y}^{-1} \sum_{i=1}^{N} y_{i}+\Sigma_{0}^{-1} \boldsymbol{\mu}_{0}\right]
$$

This is the **Maximum A-Posteriori (MAP)**.

**Derivation of MLE:**
We calculate the log likelihood of the dataset $(Y)$ given MVN parameters $(x, \sigma_{y})$:
$$
\begin{aligned}
\mathcal{L}(Y;x, \Sigma_{y} ) &=\log \prod_{i=1}^{N} p\left(y_{i} \mid x, \Sigma_{y}\right) \\
&=\frac{N}{2} \log \left|\Sigma_{y}^{-1}\right|-\frac{1}{2} \sum_{i=1}^{N} \operatorname{tr}\left[\left(y_{i}-x\right)\left(y_{i}-x\right)^{T} \Sigma_{y}^{-1}\right]
\end{aligned}
$$

Then, we set the patial derivative w.r.t $x$ equal to 0:

$$
\frac{\partial \mathcal{L}(Y;x, \Sigma_{y} )}{\partial x}=-\frac{1}{2} \sum_{i=1}^{N}-2 \Sigma_{y}^{-1}\left(y_{i}-x\right)=0$$. 

Thus the MLE is:

$$
\begin{array}{l}
\sum_{i=1}^{N}\left(y_{i}-x\right)=0 \\
\Rightarrow \quad \sum_{i=1}^{N} y_{i}-\sum_{i=1}^{N} x=0 \\
\Rightarrow \quad \mu_{M L E}=\frac{1}{N} \sum_{i=1}^{N} y_{i}
\end{array}$$

Hence, the MLE for MVN is just the empirical mean of the samples $y_{i}$. In this case, it is **the mean of sensor observations i.e. $\overline y$**.

**Part c: Plot the contour of prior and posterior distributions.**

```python
import numpy as np
import torch
from scipy.stats import multivariate_normal
from torch.distributions.multivariate_normal import MultivariateNormal as MVN
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
X,Y= np.mgrid[-2.5:2.5:.01,-2.5:2.5:0.01]
pos = np.dstack((X,Y))

def Bayes_draws(N,mult_factor,identity_size=2):
    #initialize given parameters
    mu0= torch.Tensor([0.5,0.5]).reshape((2,1))
    x= torch.Tensor([1.5,1.5]).reshape((2,1))
    sig0= mult_factor*torch.eye(identity_size)    
    sigy= torch.Tensor([[0.2, 0.1],[0.1,0.1]])
    
    #multivariate distr
    y= MVN(loc=x.T,covariance_matrix=sigy).sample((N,))
    y_bar=y.mean(dim=0).T
    
    #posteriors
    sig_n= torch.inverse(torch.inverse(sig0)+(N*torch.inverse(sigy)))
    mu_n= sig_n @ ((torch.inverse(sig0)@mu0)+(N*torch.inverse(sigy)@y_bar))
    print(mu_n)
    Prior = multivariate_normal(mean=mu0.T.reshape(2,),cov=sig0).pdf(pos)
    Posterior= multivariate_normal(mean=mu_n.T.reshape(2,),cov=sig_n).pdf(pos)
    return(Prior,Posterior)

def plot_contours(Prior,Posterior,N,mult_factor):
    fig, ax = plt.subplots(1,2,figsize=(15,7))
    
    fig.tight_layout(pad=3)
    pl1=ax[0].contourf(X,Y,Prior)
    ax[0].set_title('Prior Distribution')
    ax[0].set_xlabel('x_1') 
    ax[0].set_ylabel('x_2')
    ax[0].locator_params(nbins=10)
    divider = make_axes_locatable(ax[0])
    cax = divider.append_axes('bottom', size='5%', pad=0.5)
    fig.colorbar(pl1, extend='both', cax=cax,orientation="horizontal")
    pl2=ax[1].contourf(X,Y,Posterior)
    ax[1].set_title('Posterior Distribution')
    ax[1].set_xlabel('x_1') 
    ax[1].set_ylabel('x_2')
    ax[1].locator_params(nbins=10)
    divider = make_axes_locatable(ax[1])
    cax = divider.append_axes('bottom', size='5%', pad=0.5)
    fig.colorbar(pl2, extend='both', cax=cax,orientation="horizontal")
    
    plt.suptitle("Contour plots for N="+str(N)+" and $\Sigma_0$="+str(mult_factor)+"$I_{2}$")
    plt.show()
```

**Contour plots for $\boldsymbol{\Sigma}_0=0.1I_{2}$ and $N=10$**

```python
# for N=10 and sig_0=0.1I_2
Pr1,Post1=Bayes_draws(10,0.1)
plot_contours(Pr1,Post1,N=10,mult_factor=0.1)
```

```python
import math
import warnings
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
warnings.simplefilter('ignore', np.RankWarning)
def target_f(x):
    '''
    function to create target function
    '''
    y = np.sin(2*math.pi*x)+np.cos(4*math.pi*x)
    return y
    
def sse(y,y_hat):
    '''
    function for sum of squared estimate of errors
    '''
    sserror=np.linalg.norm(y-y_hat)**2
    return sserror

def fit_poly(M, target_f,x_train,x_test):
    '''
    function for fitting polynomial and calculating sse for train and test
    '''
    sse_test = [] 
    sse_train = []
    # Fitting polynomials with M degrees
    for m in M:
        # fit the polynomial function
        z = np.polyfit(x_train,target_f(x_train),m)
        p = np.poly1d(z)
        # get sse scores
        sse_train.append(sse(p(x_train),target_f(x_train)))
        sse_test.append(sse(p(x_test),target_f(x_test)))
        pass
    return sse_train, sse_test

def fit_poly_reg(f, m,lambdas, x_train,x_test):
    '''
    function for regularization and calculating sse for train and test
    '''
    sse_test = [] 
    sse_train = []
    X_train = np.array(x_train).reshape(-1,1)
    X_test = np.array(x_test).reshape(-1,1)
    # use lambdas for fitting polynomials
    for lambda_ in lambdas:
        # fit a ridge model
        model = make_pipeline(PolynomialFeatures(m),Ridge(alpha=(lambda_)))
        model.fit(X_train,target_f(X_train))
        # get sse scores
        sse_train.append(sse(f(X_train), model.predict(X_train)))
        sse_test.append(sse(f(X_test),model.predict(X_test)))
    return sse_train, sse_test
```

## Problem 5: Minimizing Minkowski Loss

**Part a: Show that
for $L(t, y(x)) = |t − y(x)|$, $\mathbb{E}[L]$ is minimized by the median of the conditional distribution $p(t|x)$**

We have $\mathbb{E}[L]=\iint L(t, y(x)) p(x, t)dx d t $ and $L(t, y(x))=|t-y(x)|$.

Thus,

$$\begin{array}{l}
\mathbb{E}[L]=\iint L(t, y(x)) p(x, t) d t d x  \\
\mathbb{E}[L]=\iint|t-y(x)| p(x, t) d t d x \\
\end{array}$$

Integrating over complete range:

$$\begin{array}{l}
\mathbb{E}[L]=\int\left[\int_{-\infty}^{t=y(x)}|t-y(x)| p(t \mid x) d t+\int_{t=y(x)}^{-\infty}|t-y(x)| p(t \mid x) d x\right] p(x) d t \\
\mathbb{E}[L]=\int_{-\infty}^{t=y(x)} t * p(t \mid x) d t-\int_{-\infty}^{t=y(x)} y(x) p(t \mid x) d t+\int_{t=y(x)}^{-\infty} y(x) p(t \mid x) dt-\int_{t=y(x)}^{-\infty} t * p(t \mid x) d t
\end{array}$$

Differentiating the above w.r.t $y(x)$:

$$\begin{array}{l}
\mathbb{E}[L]=0-\int_{-\infty}^{t=y(x)} 1 p(t \mid x) d t+\int_{t=y(x)}^{\infty} 1 p(t \mid x)dt-0 \\
\mathbb{E}[L]=-\int_{-\infty}^{t=y(x)} 1 p(t \mid x) d t+\int_{t=y(x)}^{\infty} 1 p(t \mid x)dt \\
\mathbb{E}[L]=0
\end{array}$$

Now; $P(t>y(x) \mid x)=P(t<y(x) \mid x)$. Thus, $\mathbb{E}[L]$ is minimized by the median of the conditional distribution of $p(t|x)$


**Alternate Proof:**

We have:
$$
\begin{aligned}
\mathbf{E}[L] &=\iint L(t, y(\mathbf{x})) p(\mathbf{x}, t) d \mathbf{x} d t \\
&=\iint|t-y(\mathbf{x})| p(\mathbf{x}, t) d \mathbf{x} d t \\
&=\int p(x) \int|t, y(\mathbf{x})| p(t \mid x) d t d \mathbf{x}
\end{aligned}
$$

Now if we reduce the above expression for all values of 't', we can reduce the loss. Let us assume M is the median.:
$$
\begin{array}
\int|t, y(\mathbf{x})| p(t \mid x) d t=\int_{-\infty}^{y(\mathbf{x})}(y(\mathbf{x})-t) p(t \mid x) d t+\int_{y(\mathbf{x})}^{\infty}(t-y(\mathbf{x})) p(t \mid x) d t \\
=\int_{-\infty}^{M}(y(\mathbf{x})-t) p(t \mid x) d t+  \int_{M}^{y(\mathbf{x})}(y(\mathbf{x})-t) p(t \mid x) d t    +\int_{y(\mathbf{x})}^{M}(t-y(\mathbf{x})) p(t \mid x) d t +\int_{M}^{\infty}(t-y(\mathbf{x})) p(t \mid x) d t \\
=\int_{-\infty}^{M}(y(\mathbf{x})-t) p(t \mid x) d t+2 \times \int_{y(\mathbf{x})}^{M}(t-y(\mathbf{x})) p(t \mid x) d t+\int_{M}^{\infty}(t-y(\mathbf{x})) p(t \mid x) d t \\
=\int_{-\infty}^{M}(y(\mathbf{x})-M+M-t) p(t \mid x) d t+2 \times \int_{y(\mathbf{x})}^{M}(t-y(\mathbf{x})) p(t \mid x) d t+\int_{M}^{\infty}(t-M+M-y(\mathbf{x})) p(t \mid x) d t
\end{array}
$$

If we split the first and the last term in the above expression and combine the terms with (M-t) and (t-M):

$$
\begin{aligned}
\int|t, y(\mathbf{x})| p(t \mid x) d t &=\int_{-\infty}^{\infty}|t-M| p(t \mid x) d t +\int_{-\infty}^{M}(y(\mathbf{x})-M) p(t \mid x) d t+\int_{M}^{\infty}(M-y(\mathbf{x})) p(t \mid x) d t+2 \times \int_{y(x)}^{M}(t-y(\mathbf{x})) p(t \mid x) d t \\
&=\mathbf{E}[|t-M|]+\int_{-\infty}^{M}(y(\mathbf{x})-M) p(t \mid x) d t+\int_{M}^{\infty}(M-y(\mathbf{x})) p(t \mid x)d t +2 \times \int_{y(x)}^{M}(t-y(\mathbf{x})) p(t \mid x) d t\\
\end{aligned}$$

Since by definition, median is the center point of a distribution, we are assuming that the distribution is equally split by the median. Substituting the two center terms accordingly:

$$
\begin{aligned}
\int|t, y(\mathbf{x})| p(t \mid x) d t&=\mathbf{E}[|t-M|]+\frac{y(\mathbf{x})-M}{2}+\frac{M-y(\mathbf{x})}{2}+2 \times \int_{y(x)}^{M}(t-y(\mathbf{x})) p(t \mid x) d t \\
&=\mathbf{E}[|t-M|]+2 \times \int_{y(x)}^{M}(t-y(\mathbf{x})) p(t \mid x) d t
\end{aligned}$$

In the above expession if we minimize the second term (with the integral), the loss will be minimized. The second term will be minimized for $y(x)=M$ as it will equate to 0. Since M is the median, we have proved that the loss is minimized by its median.

Show that $\mathbb{E}[L]$ is minimized by the mode of the conditional distribution $p(t|x)$**

```python
def gb(x,mu1,mu2):
    '''
    function for transformation using Gaussian basis function
    '''
    x1 = np.subtract(x,mu1[0])
    x2 = np.subtract(x,mu2[0])
    x1d = np.exp(-0.5*np.diagonal(np.dot(x1.T,x1)))
    x2d = np.exp(-0.5*np.diagonal(np.dot(x2.T,x2)))
    return np.vstack((x1d,x2d))

phi_x1 = gb(x_1,np.array([0,0]),np.array([-1,-1]))
phi_x2 = gb(x_2,np.array([0,0]),np.array([-1,-1]))

#plot transformed datasets
figure = plt.figure(figsize=(10, 10))
plt.scatter(phi_x1[0,:],phi_x1[1,:],c="red",label='Dataset 1 (GMM-1)')
plt.scatter(phi_x2[0,:],phi_x2[1,:],c="green",label='Dataset 2 (GMM-2)')
plt.title('GMM-1 and GMM-2 Transformed Models')
plt.xlabel('phi_1(x)') 
plt.ylabel('phi_2(y)')
plt.legend()
plt.show()
```
